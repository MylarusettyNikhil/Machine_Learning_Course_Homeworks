# Function to remove the outliers

```python
def remove_outlier(input_data, col_name):
    q1 = input_data[col_name].quantile(0.25)
    q3 = input_data[col_name].quantile(0.75)
    
   
    iqr = q3-q1 
    low_value  = q1 - 1.5*iqr      #Interquartile range
    high_value = q3 + 1.5*iqr
    
    output_data = input_data.loc[(input_data[col_name] > low_value) & (input_data[col_name] < high_value)]
    
    return output_data
```

```python
df2 = remove_outlier(df, 'housing_median_age')
df3 = remove_outlier(df2, 'total_rooms')
df4 = remove_outlier(df3, 'total_bedrooms')
df5 = remove_outlier(df4, 'population')
df6 = remove_outlier(df5, 'households')
df7 = remove_outlier(df6, 'median_income')
df8 = remove_outlier(df7, 'median_house_value')
```
